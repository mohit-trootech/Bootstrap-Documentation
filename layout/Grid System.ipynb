# Grid System

Bootstraps grid system uses a series of containers, rows, and columns to layout and align content. Its built with flexbox and is fully responsive. Below is an example and an in-depth explanation for how the grid system comes together.


    <div class="container text-center">
      <div class="row">
        <div class="col">
          Column
        </div>
        <div class="col">
          Column
        </div>
        <div class="col">
          Column
        </div>
      </div>
    </div>

## Gutters
Gutters are the padding between your columns, used to responsively space and align content in the Bootstrap grid system.
