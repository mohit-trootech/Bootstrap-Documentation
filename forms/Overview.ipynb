# Forms Overview

Bootstrap Forms Allow to Create Responsive forms with Various Customization for Each Element.

1. Form control
Style textual inputs and textareas with support for multiple states.
2. Select
Improve browser default select elements with a custom initial appearance.
3. Checks & radios
Use our custom radio buttons and checkboxes in forms for selecting input options.
4. Range
Replace browser default range inputs with our custom version.
5. Input group
Attach labels and buttons to your inputs for increased semantic value.
6. Floating labels
Create beautifully simple form labels that float over your input fields.
7. Layout
Create inline, horizontal, or complex grid-based layouts with your forms.
8. Validation
Validate your forms with custom or native validation behaviors and styles.

## First Form

As, we know that bootstrap is just using appropiate class for element. So, bootstrap provide simple class for main two element for form `label & input` which are building block for form.

- label - `form-label`
- input - `form-control`

        <form>
        <div class="mb-3">
            <label for="exampleInputEmail1" class="form-label">Email address</label>
            <input type="email" class="form-control" id="exampleInputEmail1" aria-describedby="emailHelp">
            <div id="emailHelp" class="form-text">We'll never share your email with anyone else.</div>
        </div>
        <div class="mb-3">
            <label for="exampleInputPassword1" class="form-label">Password</label>
            <input type="password" class="form-control" id="exampleInputPassword1">
        </div>
        <div class="mb-3 form-check">
            <input type="checkbox" class="form-check-input" id="exampleCheck1">
            <label class="form-check-label" for="exampleCheck1">Check me out</label>
        </div>
        <button type="submit" class="btn btn-primary">Submit</button>
        </form>

## Disabled forms 
Add the disabled boolean attribute on an input to prevent user interactions and make it appear lighter.

    <input class="form-control" id="disabledInput" type="text" placeholder="Disabled input here..." disabled>
<input class="form-control" id="disabledInput" type="text" placeholder="Disabled input here..." disabled>


## Readonly

Add the readonly boolean attribute on an input to prevent user update. user can interact with element copy the data but cannot update it.

    <input class="form-control" id="readonlyInput" type="text" placeholder="Readonly input here..." readonly>

<input class="form-control" id="readonlyInput" type="text" value="Readonly input here..." readonly>

## Sizing 

Adjust size of tags with `.form-control-lg $ .form-control-sm`

    <input class="form-control form-control-lg" type="text" placeholder=".form-control-lg" aria-label=".form-control-lg example">
    <input class="form-control" type="text" placeholder="Default input" aria-label="default input example">
    <input class="form-control form-control-sm" type="text" placeholder=".form-control-sm" aria-label=".form-control-sm example">



## Form text 

Form text allow to provide the information about the input tag. Block-level or inline-level form text can be created using `.form-text`.


    <label for="inputPassword5" class="form-label">Password</label>
    <input type="password" id="inputPassword5" class="form-control" aria-describedby="passwordHelpBlock">
    <div id="passwordHelpBlock" class="form-text">
      Your password must be 8-20 characters long, contain letters and numbers, and must not contain spaces, special characters, or emoji.
    </div>

<label for="inputPassword5" class="form-label">Password</label>
<input type="password" id="inputPassword5" class="form-control" aria-describedby="passwordHelpBlock">
<div id="passwordHelpBlock" class="form-text">
  Your password must be 8-20 characters long, contain letters and numbers, and must not contain spaces, special characters, or emoji.
</div>
